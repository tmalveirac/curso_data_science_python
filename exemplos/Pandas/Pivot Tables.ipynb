# Quantidade de Votos por Partido

```python
primary_df = pd.read_csv('primary_results.csv')
```

```python
pd.pivot_table(primary_df, index = ['party'], values = ['votes'], aggfunc = {'votes': np.sum}).apply(lambda x: x*100 / float(x.sum()))
```

# Quantidade de Votos por Candidato

```python
pd.pivot_table(primary_df, index = ['candidate'], values = ['votes'], aggfunc = {'votes': np.sum}).apply(lambda x: x*100 / float(x.sum()))['votes'].sort_values(ascending = False)
```
