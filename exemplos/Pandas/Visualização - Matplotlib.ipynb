```python
%matplotlib inline
import pandas as pd
import pydataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pydataset.data('titanic', show_doc='True')
```

```python
titanic = pydataset.data('titanic')
```

```python
titanic.head()
```

```python
titanic['class'].value_counts()
```

```python
titanic['class'].value_counts().plot(kind='bar')
plt.show() # n√£o mostrar o objeto 
```

```python
titanic['survived'].value_counts().plot(kind='bar')
plt.show()
```

```python
titanic.groupby('survived')['class'].value_counts()
```

```python
titanic.groupby('survived')['class'].value_counts().plot(kind='bar')
plt.show()
```

```python
titanic.groupby(['class', 'survived']).aggregate({'survived': 'count'}).apply(lambda x: round(x*100 / float(x.sum()), 2))
```

```python
titanic[titanic['class'] == '1st class'].groupby(['class', 'survived']).aggregate({'survived': 'count'}).apply(lambda x: round(x*100 / float(x.sum()), 2)).plot(kind='pie', subplots=True, autopct='%1.2f%%' )
```

```python
titanic[titanic['class'] == '2nd class'].groupby(['class', 'survived']).aggregate({'survived': 'count'}).apply(lambda x: round(x*100 / float(x.sum()), 2)).plot(kind='pie', subplots=True, autopct='%1.2f%%' )
```

```python
titanic[titanic['class'] == '3rd class'].groupby(['class', 'survived']).aggregate({'survived': 'count'}).apply(lambda x: round(x*100 / float(x.sum()), 2)).plot(kind='pie', subplots=True, autopct='%1.2f%%' )
```

```python
estados = pd.read_csv('populacao_brasil.csv')
```

```python
estados['total'].hist()
```

```python
fig, ax = plt.subplots()
plt.hist(estados['total'], bins=15, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
estados['percent'] = estados['total'] / estados['total'].sum()
```

```python
plt.pie(estados['percent'], labels=estados['estado'])
plt.show()
```
