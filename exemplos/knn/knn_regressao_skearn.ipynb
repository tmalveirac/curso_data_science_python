```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.DESCR
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
K = 9
knn = KNeighborsRegressor(n_neighbors=K)
knn.fit(boston.data, boston.target)
```

```python
print(boston.target[0])
print(knn.predict([boston.data[0]]))
```

```python
knn.score(boston.data, boston.target)
```

```python
y_ = knn.fit(boston.data, boston.target).predict([boston.data[12]])
y = boston.target[12]
print(y_)
print(y)
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
y_ = knn.fit(boston.data, boston.target).predict(boston.data)
y = boston.target
```

```python
plt.plot(np.linspace(-1,1,506), y, label='esperado', color='black', linewidth=0.5)
plt.plot(np.linspace(-1,1,506), y_, label='saida', color='blue', linewidth=0.5)
plt.legend()
plt.show()
```

```python
x,y = boston.data[:50], boston.target[:50]
y_ = knn.fit(x, y).predict(x)
```

```python
plt.plot(np.linspace(-1,1,50), y, label='esperado', color='black', linewidth=0.5)
plt.plot(np.linspace(-1,1,50), y_, label='saida', color='blue', linewidth=0.5)
plt.legend()
plt.show()
```

```python
#R2
knn.score(x,y)
```
