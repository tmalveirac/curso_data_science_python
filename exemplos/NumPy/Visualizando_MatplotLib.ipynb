```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
salario = np.array([100, 200, 300, 500, 400, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Black', ls='--')
```

```python
plt.plot(salario, c='Black', ls='--', marker='s')
```

```python
plt.plot(salario, c='Black', ls='--', marker='s', ms='18')
plt.show()
```

```python
salarios_m = np.array([100,200,400,300])
salarios_g = np.array([50,300, 500,450])
```

```python
plt.plot(salarios_m, c='Black', ls='--', marker='s', ms=8, label ='Salarios M' )
plt.plot(salarios_g, c='Red', ls='--', marker='s', ms=8, label ='Salarios G' )
plt.show()
```

```python
#Legenda
plt.plot(salarios_m, c='Black', ls='--', marker='s', ms=8, label ='Salarios M' )
plt.plot(salarios_g, c='Red', ls='--', marker='s', ms=8, label ='Salarios G' )
plt.legend()
plt.show()
```

```python
plt.plot(salarios_m, c='Black', ls='--', marker='s', ms=8, label ='Salarios M' )
plt.plot(salarios_g, c='Red', ls='--', marker='^', ms=8, label ='Salarios G' )
plt.legend()
plt.show()
```

```python
plt.plot(salarios_m, c='Black', ls='--', marker='s', ms=8, label ='Salarios M' )
plt.plot(salarios_g, c='Red', ls='--', marker='s', ms=8, label ='Salarios G' )
plt.legend(loc='upper right')
plt.show()
```

```python
#Colocando mais uma linha
salarios_p = [200,150,300,700]
plt.plot(salarios_m, c='Black', ls='--', marker='s', ms=10, label ='Salarios M' )
plt.plot(salarios_g, c='Red', ls='--', marker='s', ms=10, label ='Salarios G' )
plt.plot(salarios_p, c='Green', ls='--', marker='o', ms=10, label ='Salarios P' )
plt.legend(loc='upper left')
plt.show()
```
