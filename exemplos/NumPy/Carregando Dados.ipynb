# Analisando o dataset Iris

```python
data = np.genfromtxt('../iris.data', delimiter=',', usecols=(0,1,2,3))
data
```

```python
len(data)
```

```python
data[:,0] #Todos os valores da primeira coluna
```

```python
data[:50,0] #Iris Setosa
```

```python
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8)
plt.show()
```

```python
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8, label='Comp. Sepala - Iris Setosa')
plt.plot(data[50:100,0], c='Blue', ls='--', marker='o', ms=8,label='Comp. Sepala - Iris Versicolor')
plt.plot(data[100:150,0], c='Yellow', ls='--', marker='x', ms=8,label='Comp. Sepala - Iris Virginica')
plt.legend()
plt.show()
```

```python
plt.plot(data[:50,1], c='Red', ls=':', marker='s', ms=8, label='Larg. Sepala - Iris Setosa')
plt.plot(data[50:100,1], c='Blue', ls='--', marker='o', ms=8,label='Larg. Sepala - Iris Versicolor')
plt.plot(data[100:150,1], c='Yellow', ls='--', marker='x', ms=8,label='Larg. Sepala - Iris Virginica')
plt.legend()
plt.show()
```

```python
plt.plot(data[:50,2], c='Red', ls=':', marker='s', ms=8, label='Comp. Petala - Iris Setosa')
plt.plot(data[50:100,2], c='Blue', ls='--', marker='o', ms=8,label='Comp. Petala - Iris Versicolor')
plt.plot(data[100:150,2], c='Yellow', ls='--', marker='x', ms=8,label='Comp. Petala - Iris Virginica')
plt.legend()
plt.show()
```

```python
plt.plot(data[:50,3], c='Red', ls=':', marker='s', ms=8, label='Larg. Petala - Iris Setosa')
plt.plot(data[50:100,3], c='Blue', ls='--', marker='o', ms=8,label='Larg. Petala - Iris Versicolor')
plt.plot(data[100:150,3], c='Yellow', ls='--', marker='x', ms=8,label='Larg. Petala - Iris Virginica')
plt.legend()
plt.grid(linewidth=0.5)
plt.show()
```
